**DeFi Llama Yield Ranking**

DeFi Llama UI allows filtering by chain (Ethereum) and categories (Dexes, Lending), but downloaded csv does **not** filter.

Process:
- Download "yield_ranking.csv", filter dataframe by chain
- save as df_yield, select columns: chain, project (app), symbol, tvlUsd, rewardTokens, pool (address), underlyingTokens
- Download "all.csv", filter by chain & category
- save as df_all, select columnns: Unnamed: 0 (project), categories (CDP, Dexes, Lending), chain
    - only interested in how to categorize projects for now
- Join tables

**Assumption**: DeFi Llama categorization aligns with our goals/objectives

```python
import pandas as pd

df = pd.read_csv('yield_ranking.csv')

df.head(n=100)
```

```python
df.shape
```

```python
df_yield = (
    df
    .loc[df['chain'] == 'Ethereum']
)

df_yield.head(n=100)
```

```python
df_yield_1 = df_yield[['chain', 'project', 'symbol', 'tvlUsd', 'rewardTokens', 'pool', 'underlyingTokens']]

df_yield_1.head()
```

```python
df_all = pd.read_csv('all.csv', low_memory=False)

# 231,206 rows, 1401 columns
df_all.shape
```

```python
import pandas as pd

def front(self, n):
    return self.iloc[:, :n]

def back(self, n):
    return self.iloc[:, -n:]

pd.DataFrame.front = front
pd.DataFrame.back = back

df_all.back(20)

```

```python
df_all_0 = (
    df_all
    .loc[df_all['Chain'] == 'ethereum']
    .loc[(df_all['Category'] == 'Dexes') | (df_all['Category'] == 'Lending') | (df_all['Category'] == 'CDP')]
)

df_all_0.head(n=500)
```

```python
df_all_1 = df_all_0[['Unnamed: 0', 'Category', 'Chain']]

df_all_1.columns = ['project', 'category', 'chain']

list(df_all_1)
```

```python
df_all_1.head()

df_all_1.shape
```

```python
df_all_1
```

```python
df_all_2 = (
     df_all_1
        .groupby(['project', 'category', 'chain'], as_index=False)
        .count()
)

df_all_2
```

```python
list(df_all_2)
```

```python
df_all_2['chain'] = df_all_2['chain'].str.capitalize()

df_all_2
```

```python
df_all_2['project'] = df_all_2['project'].str.lower()

df_all_2
```

```python
df_all_2.loc[df_all_2['project'] == "aave v2", "project"] = "aave-v2"
df_all_2.loc[df_all_2['project'] == "iron bank", "project"] = "iron-bank"
df_all_2.loc[df_all_2['project'] == "bancor v3", "project"] = "bancor-v3"
df_all_2.loc[df_all_2['project'] == "the granary", "project"] = "the-granary"


df_all_2
```

```python
for col in df_all_2:
    print(df_all_2[col].unique())
```

```python
for col in df_yield_1:
    print(df_yield_1[col].unique())
```

```python
df_yield_1
```

```python
# left: df_yield_1
# right: df_all_2

# left merge
df_left_merge = df_yield_1.merge(df_all_2, on = ['project', 'chain'], how='left')

df_left_merge

```

```python
# inner join

df_inner_merge = df_yield_1.merge(df_all_2, on = ['project', 'chain'])

df_inner_merge
```

```python
for col in df_inner_merge:
    print(df_inner_merge[col].unique())
```
